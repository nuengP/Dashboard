```python
## import library
import pandas as pd
import numpy as np
```

```python
## load csv files
calendar = pd.read_csv("/data/notebook_files/Maven Toys Data/calendar.csv")
data_dic = pd.read_csv("/data/notebook_files/Maven Toys Data/data_dictionary.csv")
inventory = pd.read_csv("/data/notebook_files/Maven Toys Data/inventory.csv")
products = pd.read_csv("/data/notebook_files/Maven Toys Data/products.csv")
sales = pd.read_csv("/data/notebook_files/Maven Toys Data/sales.csv")
stores = pd.read_csv("/data/notebook_files/Maven Toys Data/stores.csv")
```

```python
## merge tables with left join
df = pd.merge(sales, stores, how = "left", on = "Store_ID")
df = pd.merge(df, products, how = "left", on = "Product_ID")
```

```python
## Validate Data
# checking for null values
print(f"show missing values :\n{df.isnull().sum()}")
# checking for duplicate
print(f"duplicate row : {df.duplicated().sum()}")
```
